```python
x_var=['LIMIT_BAL',
 'SEX',
 'EDUCATION',
 'MARRIAGE',
 'AGE',
 'PAY_0',
 'PAY_2',
 'PAY_3',
 'PAY_4',
 'PAY_5',
 'PAY_6',
 'BILL_AMT1',
 'BILL_AMT2',
 'BILL_AMT3',
 'BILL_AMT4',
 'BILL_AMT5',
 'BILL_AMT6',
 'PAY_AMT1',
 'PAY_AMT2',
 'PAY_AMT3',
 'PAY_AMT4',
 'PAY_AMT5',
 'PAY_AMT6',]
```

```python
x=viva[x_var]
```

```python
y=viva['default.payment.next.month']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=44)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr1=LogisticRegression()
```

```python
lr1.fit(x_train,y_train)
```

```python
y_pred=lr1.predict(x_test)
```

```python
from sklearn.metrics import recall_score,precision_score,roc_auc_score,f1_score,accuracy_score
```

```python
print(y_pred.shape)
print(y_test.shape)
```

```python
roc_auc=roc_auc_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
accuracy=accuracy_score(y_test,y_pred)
precision=precision_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
```

```python
print('roc_auc is ',roc_auc)
print('accuracy is ',accuracy)
print('precision is',precision)
print('recall is ',recall)
print('f1 score is ',f1)
```
